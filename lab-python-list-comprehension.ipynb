````python
products = ["t-shirt", "mug", "hat", "book", "keychain"]


def initialize_inventory(products):
    inventory = {product: int(input(f"Ingresa la cantidad disponible de {product}s: ")) for product in products}
    return inventory

inventory = initialize_inventory(products)

print("\nInventario inicial:")
for product, quantity in inventory.items():
    print(f"{product}: {quantity}")
````

````python
def get_customer_orders():
    num_orders = int(input("Ingresa la cantidad de pedidos de clientes: "))
    orders = [input("Ingresa el nombre de un producto que el cliente quiere ordenar: ") for _ in range(num_orders)]
    return orders

# Ejemplo de uso
customer_orders = get_customer_orders()
print("\nPedidos del cliente:")
for order in customer_orders:
    print(order)
````

````python
def calculate_total_price(customer_orders):
    prices = [float(input(f"Ingresa el precio de {product}: ")) for product in customer_orders]
    total_price = sum(prices)
    return total_price

# Ejemplo de uso con la lista de pedidos del paso anterior
total = calculate_total_price(customer_orders)
print(f"\nPrecio total: {total}")
````

````python
def update_inventory(inventory, customer_orders):
    # Restamos 1 unidad a cada producto pedido, si estÃ¡ en el inventario
    updated_inventory = {product: inventory.get(product, 0) - 1 if product in inventory else 0 for product in inventory}
    
    # Filtramos para eliminar productos con cantidad cero o menos
    updated_inventory = {product: qty for product, qty in updated_inventory.items() if qty > 0}
    
    return updated_inventory

# Ejemplo de uso
inventory = update_inventory(inventory, customer_orders)

print("\nInventario actualizado:")
for product, quantity in inventory.items():
    print(f"{product}: {quantity}")
````
